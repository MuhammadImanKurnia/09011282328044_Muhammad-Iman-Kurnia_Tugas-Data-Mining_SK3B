```python
# Mengimport pandas, numpy, matplotlib, dan seaborn (tabulate untuk membuat tabel)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
# Mengimport dataset

sns.set(style = "whitegrid")
df = pd.read_csv('C:\\Dataset\\archive\\heart_disease_uci.csv')
```

```python
# Menampilkan 5 baris pertama dari dataset

print("- 5 baris pertama : ")
df.head()
```

```python
# Menampilkan 5 baris terakhir dari dataset

print("- 5 baris terakhir : ")
df.tail()
```

```python
# Menampilkan informasi dataset

print("\n- Inforasi dataset : ")
df_info = pd.DataFrame({
    'Column': df.columns,
    'Non-Null Count': df.count().values,
    'Data Type': df.dtypes.values
})
print(tabulate(df_info, headers='keys', tablefmt='fancy_grid'))
```

```python
# Menampilkan nilai yang hilang pada dataset

missing_values = df.isnull().sum().reset_index()
missing_values.columns = ['Column', 'Missing Values']
print("\n- Nilai yang hilang : ")
print(tabulate(missing_values, headers='keys', tablefmt='fancy_grid'))
```

```python
# Box plot untuk mendeteksi outliers

plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title('Box Plot untuk Variabel Numerik')
plt.show()
```

```python
# Statistik deskriptif

print("\n- Statistik Deskriptif: \n")
print(tabulate(df.describe().reset_index(), headers='keys', tablefmt='fancy_grid'))
```

```python
# Menghitung mean, median, mode, standar deviasi, variansi, skewness, dan kurtosis

print("- Menghitung mean, median, mode, standar deviasi, variansi, skewness, dan kurtosis")
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    print(f"\nStatistik untuk {col} : ")
    print(f"Mean            : {df[col].mean()}")
    print(f"Median          : {df[col].median()}")
    print(f"Mode            : {df[col].mode()[0]}")
    print(f"Standar Deviasi : {df[col].std()}")
    print(f"Variansi        : {df[col].var()}")
    print(f"Skewness        : {df[col].skew()}")
    print(f"Kurtosis        : {df[col].kurt()}")
    print("-------------------------------------")
```

```python
# Menghitung korelasi antar variabel numerik

numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

# Membuat matriks korelasi
print("\n- Matriks Korelasi: \n")
print(tabulate(correlation_matrix.reset_index(), headers='keys', tablefmt='fancy_grid'))

# Membuat heatmap korelasi antar variabel
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

```python
# Menampilkan histogram untuk setiap variabel numerik

print("- Histogram untuk setiap variabel numerik")
df.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# Menampilkan box plot untuk setiap variabel numerik

print("- Box plot untuk setiap variabel numerik")
df.plot(kind='box', subplots=True, layout=(4,4), figsize=(15,15), sharex=False, sharey=False)
plt.show()
```

```python
# Menampilkan scatter plot untuk pasangan variabel dengan korelasi tinggi

print("- Scatter plot untuk pasangan variabel dengan korelasi tinggi")
sns.pairplot(df, hue ='age')  
plt.show()  
```

```python
# Menampilkan scatter plot untuk pasangan variabel dengan korelasi paling tinggi

# Menghitung matriks korelasi dan menemukan pasangan variabel dengan korelasi tertinggi
max_corr_pair = numeric_df.corr().abs().unstack().sort_values(ascending=False).drop_duplicates().index[1]

# Membuat scatter plot dengan garis regresi polinomial (order=2)
plt.figure(figsize=(8, 6))
sns.regplot(x=numeric_df[max_corr_pair[0]], y=numeric_df[max_corr_pair[1]], order=2, scatter_kws={'s': 50}, line_kws={"color":"red"})
plt.title(f'Scatter Plot antara {max_corr_pair[0]} dan {max_corr_pair[1]} dengan Garis Regresi Polinomial (Korelasi: {numeric_df.corr().loc[max_corr_pair[0], max_corr_pair[1]]:.2f})')
plt.xlabel(max_corr_pair[0])
plt.ylabel(max_corr_pair[1])
plt.show()
```
